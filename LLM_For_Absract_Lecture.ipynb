```python
import pandas as pd
from sentence_transformers import SentenceTransformer, util
```

Lectures des abstract pour IEEE Xplore

```python
df=pd.read_csv("IEEE_Scrapping.csv",sep=";")
```

Chargement du modèle LLM

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
similarity_scores=[]
sentence="Reinforcement learning for trading"
for i in df["Abstracts"]:
    abstract = i

    embeddings = model.encode([sentence, abstract], convert_to_tensor=True)
    similarity = util.cos_sim(embeddings[0], embeddings[1])

    print(f"Similarity score: {similarity.item():.4f}")
    similarity_scores.append(similarity.item())
```

```python
df["similarity_scores"]=similarity_scores
```

Conservation des articles ayant un score de similarité supérieur à 0.65

```python
articles_interessant=df[df["similarity_scores"]>0.65]
```

Enregistrement des articles sous CSV

```python
articles_interessant.to_csv("Meilleurs_Articles.csv",sep=";",index=False)
```
